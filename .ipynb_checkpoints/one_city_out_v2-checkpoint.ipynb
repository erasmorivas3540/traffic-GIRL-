# Protocol A-v2 – Cross-City (Mini-Batch) (Leave-One-City-Out) OOD

For each city, train on all others, validate on a 15% stratified hold-out of training cities, and test on the held-out city.

**Metrics** (mean ± std across 5 seeds): MeanEnvAcc, WorstEnvAcc, MeanEnvF1, WorstEnvF1

**Structure**
1. Non-Graph Baselines  (LR, MLP, XGBoost)
2. Graph Baselines      (GNN-ERM, GNN+GeoReg)
3. Invariant Baselines  (IRM, VREx, GroupDRO, GNN+IRM)
4. GIRL                 (proposed)
5. Ablations            (Step 5)
6. Full Comparison Table
7. Theory Alignment     (Step 7)
8. Reliability Diagrams (Step 6)

```python
import glob
import json
import os
import numpy as np
import pandas as pd
import torch
from sklearn.preprocessing import StandardScaler

from models_v2 import (
    load_data, load_graph, load_city_splits, graph_tensors,
    set_seed, fmt, summarise, fmt_city, aggregate_city_metrics,
    compute_metrics_binary, compute_metrics_gnn,
    run_lr, run_mlp, run_xgb,
    run_gnn_erm, run_gnn_georeg,
    run_irm, run_vrex, run_groupdro,
    run_gnn_irm, run_girl,
)

SEEDS  = [0, 1, 2, 3, 4]
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {DEVICE}')
```

```python
X_all, y_all = load_data(label_col='SEVERITY')

graph      = load_graph()
ei_t, ew_t = graph_tensors(graph, DEVICE)

# splits_s0 used by theory alignment cell (fold 0 representative split)
splits_s0 = load_city_splits(0)
X_np      = X_all.values   # kept for theory cell scaler
y_full    = torch.tensor(y_all, dtype=torch.long).to(DEVICE)

n_splits = len(glob.glob('splits/city_seed0_split*.json'))
print(f'Cities per seed : {n_splits}')
print(f'Features        : {X_all.shape[1]}')
print(f'Classes         : {len(np.unique(y_all))}')
print(f'Features        : {X_all.shape[1]}  (scaling handled per fold inside runners)')
```

```python
# ── Persistence helpers ────────────────────────────────────────────────────
# Save / reload aggregated city metrics so they survive kernel restarts.

def save_city_agg(model_name, city_res, protocol='one_city_out_v2'):
    """Save aggregated city metrics to disk."""
    out_dir = os.path.join('results', protocol, model_name)
    os.makedirs(out_dir, exist_ok=True)
    path = os.path.join(out_dir, 'city_aggregated.json')
    serialisable = {k: [float(v) for v in vals] for k, vals in city_res.items()}
    with open(path, 'w') as f:
        json.dump(serialisable, f, indent=2)
    print(f'  Saved: {path}')


def reload_city_agg(model_name, protocol='one_city_out_v2'):
    """Reload aggregated city metrics saved by save_city_agg."""
    path = os.path.join('results', protocol, model_name, 'city_aggregated.json')
    if not os.path.exists(path):
        print(f'  Missing: {path} — re-run the training loop')
        return {}
    data = json.load(open(path))
    print(f'  Loaded: {path}  ({len(data.get("mean_f1",[]))} seeds)')
    return data


def city_agg_exists(model_name, protocol='one_city_out_v2'):
    path = os.path.join('results', protocol, model_name, 'city_aggregated.json')
    return os.path.exists(path)


print('Persistence helpers defined')
```

## 1. Non-Graph Baselines

### 1.1 Logistic Regression

```python
if city_agg_exists('logistic_regression'):
    print('Loading saved LR results...')
    LR_city = reload_city_agg('logistic_regression')
else:
    def run_city_lr(seeds):
        city_res = {'mean_acc':[], 'worst_acc':[], 'mean_f1':[], 'worst_f1':[],
                    'mean_bal_acc':[], 'worst_bal_acc':[]}
        for seed in seeds:
            splits = load_city_splits(seed)
            env_metrics = []
            for sp in splits:
                ti, vi, tei = sp['train_idx'], sp['val_idx'], sp['test_idx']
                r = run_lr(X_all.iloc[ti], y_all[ti],
                           X_all.iloc[vi], y_all[vi],
                           X_all.iloc[tei], y_all[tei],
                           seed=seed, protocol='one_city_out_v2')
                env_metrics.append({k: r[k][-1] for k in ['acc','macro_f1','bal_acc']})
            agg = aggregate_city_metrics(env_metrics)
            for k in city_res: city_res[k].append(agg[k])
            print(f'  Seed {seed}: MeanF1={agg["mean_f1"]:.4f} WorstF1={agg["worst_f1"]:.4f}')
        return city_res

    print('Logistic Regression – Cross-City')
    LR_city = run_city_lr(SEEDS)
    save_city_agg('logistic_regression', LR_city)

fmt_city(LR_city)
```

### 1.2 MLP (ERM)

```python
if city_agg_exists('mlp'):
    print('Loading saved MLP results...')
    MLP_city = reload_city_agg('mlp')
else:
    def run_city_mlp(seeds):
        city_res = {'mean_acc':[], 'worst_acc':[], 'mean_f1':[], 'worst_f1':[],
                    'mean_bal_acc':[], 'worst_bal_acc':[]}
        for seed in seeds:
            splits = load_city_splits(seed)
            env_metrics = []
            for sp in splits:
                ti, vi, tei = sp['train_idx'], sp['val_idx'], sp['test_idx']
                r = run_mlp(X_all.iloc[ti], y_all[ti],
                            X_all.iloc[vi], y_all[vi],
                            X_all.iloc[tei], y_all[tei],
                            seed=seed, device=DEVICE, protocol='one_city_out_v2')
                env_metrics.append({k: r[k][-1] for k in ['acc','macro_f1','bal_acc']})
            agg = aggregate_city_metrics(env_metrics)
            for k in city_res: city_res[k].append(agg[k])
            print(f'  Seed {seed}: MeanF1={agg["mean_f1"]:.4f} WorstF1={agg["worst_f1"]:.4f}')
        return city_res

    print('MLP – Cross-City')
    MLP_city = run_city_mlp(SEEDS)
    save_city_agg('mlp', MLP_city)

fmt_city(MLP_city)
```

### 1.3 XGBoost

```python
if city_agg_exists('xgboost'):
    print('Loading saved XGB results...')
    XGB_city = reload_city_agg('xgboost')
else:
    def run_city_xgb(seeds):
        city_res = {'mean_acc':[], 'worst_acc':[], 'mean_f1':[], 'worst_f1':[],
                    'mean_bal_acc':[], 'worst_bal_acc':[]}
        for seed in seeds:
            splits = load_city_splits(seed)
            env_metrics = []
            for sp in splits:
                ti, vi, tei = sp['train_idx'], sp['val_idx'], sp['test_idx']
                r = run_xgb(X_all.iloc[ti], y_all[ti],
                            X_all.iloc[vi], y_all[vi],
                            X_all.iloc[tei], y_all[tei],
                            seed=seed, protocol='one_city_out_v2')
                env_metrics.append({k: r[k][-1] for k in ['acc','macro_f1','bal_acc']})
            agg = aggregate_city_metrics(env_metrics)
            for k in city_res: city_res[k].append(agg[k])
            print(f'  Seed {seed}: MeanF1={agg["mean_f1"]:.4f} WorstF1={agg["worst_f1"]:.4f}')
        return city_res

    print('XGBoost – Cross-City')
    XGB_city = run_city_xgb(SEEDS)
    save_city_agg('xgboost', XGB_city)

fmt_city(XGB_city)
```

## 2. Graph Baselines

> **Important:** GNN models refit a StandardScaler on each fold's train nodes before building `x_fold`. This ensures no feature leakage from val/test nodes into the scaler, and gives each city fold the correct feature normalisation.

### 2.1 GNN – ERM

```python
if city_agg_exists('gnn_erm'):
    print('Loading saved GNN-ERM results...')
    GNN_ERM_city = reload_city_agg('gnn_erm')
else:
    def run_city_gnn_erm(seeds):
        city_res = {'mean_acc':[], 'worst_acc':[], 'mean_f1':[], 'worst_f1':[],
                    'mean_bal_acc':[], 'worst_bal_acc':[]}
        for seed in seeds:
            splits = load_city_splits(seed)
            env_metrics = []
            for sp in splits:
                ti = np.array(sp['train_idx'])
                vi, tei = sp['val_idx'], sp['test_idx']
                # Refit scaler on this fold's train nodes only
                torch.cuda.empty_cache()
                r = run_gnn_erm(X_all, y_full, ei_t, ti, vi, tei,
                                seed=seed, epochs=200, patience=20, device=DEVICE,
                                protocol='one_city_out_v2')
                env_metrics.append({k: r[k][-1] for k in ['acc','macro_f1','bal_acc']})
            agg = aggregate_city_metrics(env_metrics)
            for k in city_res: city_res[k].append(agg[k])
            print(f'  Seed {seed}: MeanF1={agg["mean_f1"]:.4f} WorstF1={agg["worst_f1"]:.4f}')
        return city_res

    print('GNN-ERM – Cross-City')
    GNN_ERM_city = run_city_gnn_erm(SEEDS)
    save_city_agg('gnn_erm', GNN_ERM_city)

fmt_city(GNN_ERM_city)
```

### 3.4 GNN + IRM

```python
if city_agg_exists('gnn_irm'):
    print('Loading saved GNN+IRM results...')
    GNN_IRM_city = reload_city_agg('gnn_irm')
else:
    def run_city_gnn_irm(seeds):
        city_res = {'mean_acc':[], 'worst_acc':[], 'mean_f1':[], 'worst_f1':[],
                    'mean_bal_acc':[], 'worst_bal_acc':[]}
        for seed in seeds:
            splits = load_city_splits(seed)
            env_metrics = []
            for sp in splits:
                ti = np.array(sp['train_idx'])
                vi, tei = sp['val_idx'], sp['test_idx']
                torch.cuda.empty_cache()
                n = len(ti); sz = n // 3
                envs = [ti[:sz], ti[sz:2*sz], ti[2*sz:]]
                r = run_gnn_irm(X_all, y_full, ei_t, ti, vi, tei,
                                seed=seed, node_envs=envs,
                                epochs=200, patience=20, device=DEVICE,
                                protocol='one_city_out_v2')
                env_metrics.append({k: r[k][-1] for k in ['acc','macro_f1','bal_acc']})
            agg = aggregate_city_metrics(env_metrics)
            for k in city_res: city_res[k].append(agg[k])
            print(f'  Seed {seed}: MeanF1={agg["mean_f1"]:.4f} WorstF1={agg["worst_f1"]:.4f}')
        return city_res

    print('GNN + IRM – Cross-City')
    GNN_IRM_city = run_city_gnn_irm(SEEDS)
    save_city_agg('gnn_irm', GNN_IRM_city)

fmt_city(GNN_IRM_city)
```

## 4. GIRL – Proposed Method

```python
# ── GIRL – Full (λ>0, μ>0) ─────────────────────────────────────────────────
if city_agg_exists('girl'):
    print('Loading saved GIRL results...')
    GIRL_city = reload_city_agg('girl')
else:
    def run_city_girl(seeds):
        city_res = {'mean_acc':[], 'worst_acc':[], 'mean_f1':[], 'worst_f1':[],
                    'mean_bal_acc':[], 'worst_bal_acc':[]}
        for seed in seeds:
            splits = load_city_splits(seed)
            env_metrics = []
            for sp in splits:
                ti = np.array(sp['train_idx'])
                vi, tei = sp['val_idx'], sp['test_idx']
                torch.cuda.empty_cache()
                n = len(ti); sz = n // 3
                envs = [ti[:sz], ti[sz:2*sz], ti[2*sz:]]
                r = run_girl(
                    X_all, y_full, ei_t, ew_t, ti, vi, tei,
                    seed=seed, node_envs=envs,
                    lam_grid=[0.1, 1.0],
                    mu_grid =[0.0001, 0.001, 0.01],
                    epochs=200, patience=20, device=DEVICE,
                    protocol='one_city_out_v2')
                env_metrics.append({k: r[k][-1] for k in ['acc','macro_f1','bal_acc']})
            agg = aggregate_city_metrics(env_metrics)
            for k in city_res: city_res[k].append(agg[k])
            print(f'  Seed {seed}: MeanF1={agg["mean_f1"]:.4f} WorstF1={agg["worst_f1"]:.4f}')
        return city_res

    print('GIRL – Cross-City')
    GIRL_city = run_city_girl(SEEDS)
    save_city_agg('girl', GIRL_city)

fmt_city(GIRL_city)
```

## 7. Theory Alignment (Step 7)

```python
# ── Step 7: Theory Alignment – Cross-City OOD ─────────────────────────────
import torch
import numpy as np
import matplotlib
matplotlib.rcParams.update({'figure.facecolor': 'white',
                             'axes.facecolor':  'white',
                             'savefig.facecolor':'white'})
import matplotlib.pyplot as plt
from scipy.sparse import csr_matrix
from sklearn.preprocessing import StandardScaler
from models_v2 import load_checkpoint, GNN, GIRL as GIRLModel
import json

PROTOCOL = 'one_city_out_v2'
SEED     = 0

# Use first city fold as representative split
sp0       = splits_s0[0]
train_idx = np.array(sp0['train_idx'])
val_idx   = sp0['val_idx']
test_idx  = sp0['test_idx']

# Refit scaler for this fold
scaler      = StandardScaler().fit(X_np[train_idx])
X_scaled    = scaler.transform(X_np)
x_full_fold = torch.tensor(X_scaled, dtype=torch.float32).to(DEVICE)

print(f'Using city fold 0  |  train={len(train_idx)}  val={len(val_idx)}  test={len(test_idx)}')


def build_laplacian(edge_index, edge_weight, n_nodes):
    ei  = edge_index.cpu().numpy()
    ew  = edge_weight.cpu().numpy() if edge_weight is not None else np.ones(ei.shape[1])
    row, col = ei[0], ei[1]
    A        = csr_matrix((ew, (row, col)), shape=(n_nodes, n_nodes))
    A        = (A + A.T) / 2
    deg      = np.array(A.sum(axis=1)).flatten()
    deg_inv  = np.where(deg > 0, deg ** -0.5, 0.0)
    D_inv    = csr_matrix((deg_inv, (np.arange(n_nodes), np.arange(n_nodes))),
                           shape=(n_nodes, n_nodes))
    return csr_matrix(np.eye(n_nodes)) - D_inv @ A @ D_inv


def dirichlet_energy(Z, L):
    return float(np.trace(Z.T @ (L @ Z)))


def compute_delta_g(edge_index, edge_weight, train_idx, test_idx, n_nodes):
    ei  = edge_index.cpu().numpy()
    ew  = edge_weight.cpu().numpy() if edge_weight is not None else np.ones(ei.shape[1])
    deg = np.zeros(n_nodes)
    np.add.at(deg, ei[0], ew)
    mean_train = deg[np.array(train_idx)].mean()
    mean_test  = deg[np.array(test_idx)].mean()
    mean_all   = deg.mean()
    return float(abs(mean_train - mean_test) / (mean_all + 1e-8)), float(mean_train), float(mean_test)


n_nodes = X_all.shape[0]
delta_g, deg_train, deg_test = compute_delta_g(ei_t, ew_t, train_idx, test_idx, n_nodes)

print(f'\nΔ_G = {delta_g:.4f}  (train deg={deg_train:.2f}, test deg={deg_test:.2f})')

print('Building Laplacian...')
L = build_laplacian(ei_t, ew_t, n_nodes)

n_classes  = int(y_full.max().item()) + 1
gnn_kwargs = {'in_dim': x_full_fold.shape[1], 'hidden_dim': 64, 'out_dim': n_classes}

girl_log = json.load(open(f'results/{PROTOCOL}/girl/seed_{SEED}.json'))
best_lam  = girl_log['hyperparameters'].get('best_lambda', 1.0)
best_mu   = girl_log['hyperparameters'].get('best_mu', 0.1)

state_erm = load_checkpoint(PROTOCOL, 'gnn_erm', SEED)
gnn_erm   = GNN(**gnn_kwargs).to(DEVICE)
gnn_erm.load_state_dict(state_erm); gnn_erm.eval()
with torch.no_grad():
    Z_erm = gnn_erm.encode(x_full_fold, ei_t, ew_t).cpu().numpy()

state_girl = load_checkpoint(PROTOCOL, 'girl', SEED)
girl_m     = GIRLModel(**gnn_kwargs).to(DEVICE)
girl_m.load_state_dict(state_girl); girl_m.eval()
with torch.no_grad():
    Z_girl = girl_m.encode(x_full_fold, ei_t, ew_t).cpu().numpy()

energy_erm  = dirichlet_energy(Z_erm,  L)
energy_girl = dirichlet_energy(Z_girl, L)
reduction   = (energy_erm - energy_girl) / (energy_erm + 1e-8) * 100

print(f'GNN-ERM  tr(Zᵀ L Z) = {energy_erm:.2f}')
print(f'GIRL     tr(Zᵀ L Z) = {energy_girl:.2f}')
print(f'GIRL reduces smoothness by {reduction:.1f}%')

erm_f1 = float(np.mean(GNN_ERM_city['mean_f1']))
drops  = {}
models_to_compare = {
    'GNN-ERM':    GNN_ERM_city,
    'GNN+GeoReg': GNN_GEO_city,
    'GNN+IRM':    GNN_IRM_city,
    'GIRL':       GIRL_city,
}
for name, res in models_to_compare.items():
    f1_mean = float(np.mean(res['mean_f1']))
    drops[name] = erm_f1 - f1_mean

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

ax = axes[0]
bars = ax.bar(['GNN-ERM', 'GIRL'], [energy_erm, energy_girl],
              color=['#5b9bd5', '#ed7d31'], alpha=0.85, width=0.4)
ax.set_ylabel('Dirichlet Energy  tr(Zᵀ L Z)', fontsize=11)
ax.set_title('Representation Smoothness\n(lower = more geographically consistent)', fontsize=11)
for bar, val in zip(bars, [energy_erm, energy_girl]):
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + max(energy_erm, energy_girl)*0.01,
            f'{val:.1f}', ha='center', va='bottom', fontsize=10, fontweight='bold')
ax.set_ylim(0, max(energy_erm, energy_girl) * 1.2)
ax.spines[['top','right']].set_visible(False)

ax = axes[1]
names  = list(drops.keys())
values = [drops[n] for n in names]
colors_drop = ['#5b9bd5' if v >= 0 else '#ed7d31' for v in values]
bars = ax.bar(names, values, color=colors_drop, alpha=0.85, width=0.5)
ax.axhline(0, color='black', lw=1, ls='--')
ax.set_ylabel('MeanEnvF1 Drop vs GNN-ERM\n(negative = improvement)', fontsize=11)
ax.set_title('OOD Performance Drop\n(lower = better generalisation)', fontsize=11)
for bar, val in zip(bars, values):
    ypos = val + 0.002 if val >= 0 else val - 0.008
    ax.text(bar.get_x() + bar.get_width()/2, ypos,
            f'{val:+.4f}', ha='center', va='bottom', fontsize=9, fontweight='bold')
ax.spines[['top','right']].set_visible(False)
ax.tick_params(axis='x', rotation=15)

plt.suptitle('Step 7: Theory Alignment – Cross-City OOD', fontsize=13, y=1.02)
plt.tight_layout()
plt.savefig('theory_alignment_one_city_out.pdf', bbox_inches='tight', facecolor='white', dpi=150)
plt.show()

print(f'\nΔ_G = {delta_g:.4f}')
print(f'GNN-ERM smoothness = {energy_erm:.2f}')
print(f'GIRL    smoothness = {energy_girl:.2f}')
print(f'Reduction          = {reduction:.1f}%')
# Replace the last two print lines with:
ok1 = '✅' if energy_girl < energy_erm else '❌'
ok2 = '✅' if drops.get('GIRL', 0) <= 0 else '❌'
print(f"{ok1}  GIRL smoother than GNN-ERM")
print(f"{ok2}  GIRL ≥ GNN-ERM on Macro-F1")
```

```python
import matplotlib.pyplot as plt
import matplotlib.backends.backend_pdf as pdf_backend
import pandas as pd
import numpy as np

# ── 1. Helpers ────────────────────────────────────────────────────────────────

def fmt(vals):
    if not vals or len(vals) == 0: return '—'
    return f'{np.mean(vals):.4f} ± {np.std(vals):.4f}'

def make_table_fig(df, title, figsize=None):
    nrows, ncols = df.shape
    if figsize is None:
        figsize = (max(12, ncols * 2.8), max(2, nrows * 0.6 + 1.5))
    fig, ax = plt.subplots(figsize=figsize, facecolor='white')
    ax.set_facecolor('white'); ax.axis('off')
    tbl = ax.table(cellText=df.values.tolist(), rowLabels=list(df.index),
                   colLabels=list(df.columns), loc='center', cellLoc='center')
    tbl.auto_set_font_size(False); tbl.set_fontsize(9)
    tbl.auto_set_column_width(col=list(range(len(df.columns))))
    for j in range(len(df.columns)):
        tbl[0, j].set_facecolor('#2c5f8a')
        tbl[0, j].set_text_props(color='white', fontweight='bold')
    for i in range(len(df.index)):
        tbl[i + 1, -1].set_facecolor('#eef3f8')
        for j in range(len(df.columns)):
            tbl[i + 1, j].set_facecolor('#f7fafd' if i % 2 == 0 else 'white')
    ax.set_title(title, fontsize=11, fontweight='bold', pad=16, color='#1a1a2e')
    plt.tight_layout()
    return fig

# ── 2. Build tables ───────────────────────────────────────────────────────────

METRICS = ['mean_acc', 'worst_acc', 'mean_f1', 'worst_f1']
COLS    = ['MeanEnvAcc', 'WorstEnvAcc', 'MeanEnvF1', 'WorstEnvF1']

rows = []
for name, res in [
    ('Logistic Regression', LR_city),
    ('MLP (ERM)',           MLP_city),
    ('XGBoost',             XGB_city),
    ('GNN-ERM',             GNN_ERM_city),
    ('GNN + GeoReg',        GNN_GEO_city),
    ('IRM',                 IRM_city),
    ('VREx',                VREX_city),
    ('GroupDRO',            DRO_city),
    ('GNN + IRM',           GNN_IRM_city),
    ('GIRL (proposed)',     GIRL_city),
]:
    row = {'Model': name}
    for m, col in zip(METRICS, COLS):
        row[col] = fmt(res.get(m, []))
    rows.append(row)
df_main = pd.DataFrame(rows).set_index('Model')

abl_rows = []
for name, res in [
    ('GNN-ERM (baseline)',             GNN_ERM_city),
    ('GIRL – IRM only (λ>0, μ=0)',    GIRL_IRM_only_city),
    ('GIRL – GeoReg only (λ=0, μ>0)', GIRL_GEO_only_city),
    ('GIRL – Full (λ>0, μ>0)',         GIRL_city),
]:
    row = {'Model': name}
    for m, col in zip(METRICS, COLS):
        row[col] = fmt(res.get(m, []))
    abl_rows.append(row)
df_abl = pd.DataFrame(abl_rows).set_index('Model')

# ── 3. Theory alignment figure (Section 7) ───────────────────────────────────

erm_f1 = float(np.mean(GNN_ERM_city['mean_f1']))
drops  = {name: erm_f1 - float(np.mean(res['mean_f1']))
          for name, res in [('GNN-ERM', GNN_ERM_city), ('GNN+GeoReg', GNN_GEO_city),
                             ('GNN+IRM', GNN_IRM_city), ('GIRL', GIRL_city)]}

fig_theory, axes = plt.subplots(1, 2, figsize=(12, 5), facecolor='white')
ax = axes[0]
bars = ax.bar(['GNN-ERM', 'GIRL'], [energy_erm, energy_girl],
              color=['#5b9bd5', '#ed7d31'], alpha=0.85, width=0.4)
ax.set_ylabel('Dirichlet Energy  tr(Zᵀ L Z)', fontsize=11)
ax.set_title('Representation Smoothness\n(lower = more geographically consistent)', fontsize=11)
for bar, val in zip(bars, [energy_erm, energy_girl]):
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + max(energy_erm, energy_girl) * 0.01,
            f'{val:.1f}', ha='center', va='bottom', fontsize=10, fontweight='bold')
ax.set_ylim(0, max(energy_erm, energy_girl) * 1.2)
ax.spines[['top', 'right']].set_visible(False)

ax = axes[1]
names  = list(drops.keys())
values = [drops[n] for n in names]
ax.bar(names, values, color=['#5b9bd5' if v >= 0 else '#ed7d31' for v in values],
       alpha=0.85, width=0.5)
ax.axhline(0, color='black', lw=1, ls='--')
ax.set_ylabel('MeanEnvF1 Drop vs GNN-ERM\n(negative = improvement)', fontsize=11)
ax.set_title('OOD Performance Drop\n(lower = better generalisation)', fontsize=11)
for i, (name, val) in enumerate(zip(names, values)):
    ax.text(i, val + 0.002 if val >= 0 else val - 0.008,
            f'{val:+.4f}', ha='center', va='bottom', fontsize=9, fontweight='bold')
ax.spines[['top', 'right']].set_visible(False)
ax.tick_params(axis='x', rotation=15)
fig_theory.suptitle('Step 7: Theory Alignment – Cross-City OOD', fontsize=13, y=1.02)
plt.tight_layout()

# ── 4. Reliability diagrams figure (Section 8) ───────────────────────────────

fig_reliability, axes = plt.subplots(2, 5, figsize=(25, 10), facecolor='white')
axes = axes.flatten()
for ax, (name, probs) in zip(axes, all_probs.items()):
    bin_accs, bin_confs, bin_counts = compute_reliability_bins(yte_city, probs, n_bins=10)
    ax.bar(bin_confs, bin_accs, width=0.08, alpha=0.75, color='steelblue', label='Model')
    ax.plot([0, 1], [0, 1], 'k--', lw=1.5, label='Perfect')
    ax.set_xlim(0, 1); ax.set_ylim(0, 1)
    ax.set_xlabel('Confidence'); ax.set_ylabel('Accuracy')
    ece_val = float(np.sum(np.abs(bin_accs - bin_confs) * bin_counts / max(bin_counts.sum(), 1)))
    ax.set_title(f'{name}\nECE={ece_val:.3f}')
    ax.legend(fontsize=7)
for ax in axes[len(all_probs):]:
    ax.set_visible(False)
fig_reliability.suptitle('Reliability Diagrams – Cross-City OOD\n(pooled across all cities, seed 0)',
                          fontsize=13, y=1.01)
plt.tight_layout()

# ── 5. Write everything to one PDF ───────────────────────────────────────────

pdf_path = 'results_one_city_out_v2.pdf'
with pdf_backend.PdfPages(pdf_path) as pdf:

    # Page 1 – Main results table
    fig = make_table_fig(
        df_main,
        'Table – Cross-City OOD Results  (mean ± std, 5 seeds)\n'
        'Leave-one-city-out evaluation across 25 cities',
        figsize=(16, 6),
    )
    pdf.savefig(fig, bbox_inches='tight'); plt.close(fig)

    # Page 2 – Ablation table
    fig = make_table_fig(
        df_abl,
        'Ablation – Cross-City OOD  (mean ± std, 5 seeds)\n'
        'Effect of IRM penalty (λ) and GeoReg penalty (μ)',
        figsize=(16, 4),
    )
    pdf.savefig(fig, bbox_inches='tight'); plt.close(fig)

    # Page 3 – Theory alignment
    pdf.savefig(fig_theory, bbox_inches='tight'); plt.close(fig_theory)

    # Page 4 – Reliability diagrams
    pdf.savefig(fig_reliability, bbox_inches='tight'); plt.close(fig_reliability)

    pdf.infodict()['Title']   = 'GIRL – Cross-City OOD Results'
    pdf.infodict()['Subject'] = 'Protocol A: Leave-one-city-out'

print(f'✅ Saved: {pdf_path}  (4 pages)')
print(df_main.to_string())
```

## Summary — Cross-City OOD Protocol (Protocol A)

**Split:** Leave-one-city-out — 25 city folds × 5 seeds | **Metrics:** MeanEnvAcc, WorstEnvAcc, MeanEnvF1, WorstEnvF1 (mean ± std)

On the cross-city protocol, **GIRL-Full achieves the best WorstEnvAcc (0.491 ± 0.081) and WorstEnvF1 (0.414 ± 0.031)** among all graph models, and matches GNN+IRM on WorstEnvBal (0.497), confirming that the joint objective improves worst-case city robustness compared to GNN-ERM (WorstEnvF1 0.411) and both ablations — IRM-only (0.374) and GeoReg-only (0.395) both underperform GIRL-Full on worst-env, making this the one protocol where the full objective clearly outperforms its components. Mean metrics are tightly clustered across all graph models (~0.765 MeanAcc, ~0.672–0.679 MeanF1), reflecting that mean performance is similar but GIRL differentiates on tail robustness. Non-graph baselines are competitive on mean F1 (XGBoost 0.688) but collapse on worst-env (LR WorstF1 0.228, GroupDRO 0.347), highlighting the importance of graph structure for city-level robustness. The theory alignment result is anomalous — Δ_G = 0.259 (the largest across all three protocols, as expected for cross-city shift), but GIRL's representations are *less* smooth than GNN-ERM (tr(Z^T L Z) increased by 162%), suggesting the IRM penalty is dominating the GeoReg constraint at the selected hyperparameters on fold 0 — this warrants investigation before submission, either by re-examining the (λ, μ) balance on the cross-city protocol specifically or by reporting the theory alignment result on a more representative fold.
