```python
def is_subset_fast(df_small, df_large):
    small_set = set(map(tuple, df_small.values))
    large_set = set(map(tuple, df_large.values))
    return small_set.issubset(large_set)

print("Subset check:", is_subset_fast(df_school_zone, df_all_sb_accidents))
```

```python
import pandas as pd

# Load datasets
all_sb = pd.read_csv("all_sb_accidents.csv", low_memory=False)
school_zone = pd.read_csv("school_zone.csv", low_memory=False)
outside_sz = pd.read_csv("outisde_sz.csv", low_memory=False)

# Ensure CASE_ID is treated consistently
for df in [all_sb, school_zone, outside_sz]:
    df["CASE_ID"] = df["CASE_ID"].astype(str)

# Convert to sets
all_ids = set(all_sb["CASE_ID"])
school_ids = set(school_zone["CASE_ID"])
outside_ids = set(outside_sz["CASE_ID"])

# Basic counts
print("Row Counts:")
print(f"all_sb_accidents: {len(all_sb)}")
print(f"school_zone:      {len(school_zone)}")
print(f"outside_sz:       {len(outside_sz)}")
print()

print("Unique CASE_ID Counts:")
print(f"all_sb_accidents: {len(all_ids)}")
print(f"school_zone:      {len(school_ids)}")
print(f"outside_sz:       {len(outside_ids)}")
print()

# Overlap analysis
overlap_school_outside = school_ids.intersection(outside_ids)
school_not_in_all = school_ids - all_ids
outside_not_in_all = outside_ids - all_ids

print("Overlap Analysis:")
print(f"School ∩ Outside: {len(overlap_school_outside)}")
print(f"School not in all_sb: {len(school_not_in_all)}")
print(f"Outside not in all_sb: {len(outside_not_in_all)}")
print()

# Optional: show example overlapping IDs
print("Example overlapping CASE_IDs:")
print(list(overlap_school_outside)[:10])
```

```python
import pandas as pd

# Load datasets
school_zone = pd.read_csv("school_zone.csv", low_memory=False)
outside_sz = pd.read_csv("outisde_sz.csv", low_memory=False)

# Ensure consistent type
school_zone["CASE_ID"] = school_zone["CASE_ID"].astype(str)
outside_sz["CASE_ID"] = outside_sz["CASE_ID"].astype(str)

# Create unique ID sets
school_ids = set(school_zone["CASE_ID"])
outside_ids = set(outside_sz["CASE_ID"])

# Check subset condition
is_subset = school_ids.issubset(outside_ids)

print("Unique school_zone CASE_IDs:", len(school_ids))
print("Unique outside_sz CASE_IDs:", len(outside_ids))
print("Intersection size:", len(school_ids.intersection(outside_ids)))
print("Is school_zone a subset of outside_sz?")
print(is_subset)

# If not subset, show missing IDs
if not is_subset:
    missing_ids = school_ids - outside_ids
    print("Number of school_zone IDs not in outside_sz:", len(missing_ids))
    print("Example missing IDs:", list(missing_ids)[:10])
```

```python
files = {
    "all_sb_accidents": "all_sb_accidents.csv",
    "school_zone": "school_zone.csv",
    "outside_sz": "outisde_sz.csv"
}

for name, path in files.items():
    df = pd.read_csv(path, low_memory=False)
    df["CASE_ID"] = df["CASE_ID"].astype(str)
    
    print(f"\n{name}")
    print("Total rows:", len(df))
    print("Unique CASE_IDs:", df["CASE_ID"].nunique())
    print("Duplicates:", len(df) - df["CASE_ID"].nunique())
```
